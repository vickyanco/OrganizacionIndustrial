```python
# consigna
# tomo muestras de la muestra x, con n = 1000
# matriz de 100000 filas por n columnas
# graficos de media y desvio estandar
```

```python
# importo librerias a utilizar
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 10, 1 #defino variables
x = np.random.normal(mu, sigma, size = 100000) #defino dist. normal
```

```python
media_de_medias = [] #lista vacia
desvios = [] #lista vacia

for i in range(0,1000):

  medias = [] #lista vacia

  for j in range(0,100):
    j += 1 #itero el contador j
    l = np.random.choice(x, size=i) #tomo muestas de x (1000)
    a = np.mean(l) #calculo la media de esas muestras
    medias.append(a) #agrego las medias a mi lista medias
  
  i += 1 #itero el contador i

  b = np.mean(medias) #calculo la media de esas muestras
  media_de_medias.append(b) #agrego las medias a mi lista media de medias

  c = np.std(medias) #calculo el desvio standar de esas muestras
  desvios.append(c) #agrego los desvios a mi lista desvios
```

```python
# grafico de media
plt.plot(media_de_medias) 
plt.show()
```

```python
# grafico de desvio estandar
plt.plot(desvios) 
plt.show()
```
